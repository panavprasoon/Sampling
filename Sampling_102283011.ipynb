```python
#Oversampling to handle imbalanced data
```

```python
X = data.drop('Class',axis=1)
y = data['Class']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_smote,y_smote = SMOTE().fit_resample(X,y)
```

```python
#After Oversampling
y_smote.value_counts()
```

```python
data_smote = pd.concat([pd.DataFrame(X_smote, columns=X.columns), pd.DataFrame(y_smote, columns=['Class'])], axis=1)
```

```python
data_smote.head()
```

```python
X_smote1=data_smote.drop("Class",axis=1)
y_smote1=data_smote["Class"]
```

```python
import numpy as np
```

```python
#1.Simple Random Sampling
```

```python
Z= 1.96
p= 0.5
E= 0.05
```

```python
n= ((Z**2) * p * (1 - p)) / (E**2)
```

```python
n=int(np.ceil(n))
```

```python
n
```

```python
from sklearn.utils import resample
sample1 = resample(data_smote, replace=False, n_samples=n)
print(sample1['Class'].value_counts())
```

```python
#2.Systematic Sampling
```

```python
N = len(data_smote)
k = N // n
if N % n != 0:
  interval = k + 1
else:
  interval= k
start = np.random.randint(interval)
indexes = range(start, N, interval)
sample2 = data_smote.iloc[indexes]
print(sample2['Class'].value_counts())
```

```python
#3.Stratified Sampling
```

```python
n=392
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
stratified_sampler = StratifiedShuffleSplit(n_splits=1, test_size=n)
for sample_indices, _ in stratified_sampler.split(X_smote, y_smote):
    x_sample = X_smote.loc[sample_indices]
    y_sample = y_smote.loc[sample_indices]
sample3 = pd.concat([pd.DataFrame(x_sample, columns=X.columns), pd.DataFrame(y_sample, columns=['Class'])], axis=1)
sample3['Class'].value_counts()
```

```python
#4.Cluster Sampling
```

```python
from sklearn.cluster import KMeans
import random
```

```python
num_clusters = 5

kmeans = KMeans(n_clusters=num_clusters, n_init='auto', random_state=42)

clusters = kmeans.fit_predict(data_smote)
clusters = pd.Series(clusters)

selected_clusters = random.sample(range(num_clusters), 3)
sample4 = data_smote.loc[clusters.isin(selected_clusters)]
sample4
```

```python
#5.Bootstrap Sampling
```

```python
n_bootstrap = 100
BootstrapSample = data_smote.copy()
sample5 = []
for _ in range(n_bootstrap):
    resampled_data = BootstrapSample.sample(n=len(data_smote), replace=True, random_state=42)
    sample5.append(resampled_data)
bootstrap_means = [resampled_data.mean(axis=0) for resampled_data in sample5]
mean_estimate = np.mean(bootstrap_means, axis=0)  # Overall mean estimate
confidence_intervals = np.percentile(bootstrap_means, [2.5, 97.5], axis=0)  # 95% CIs
sample5
```
